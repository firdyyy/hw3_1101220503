```python
import pandas as pd
```

# Membaca Data CSV
- Membaca file CSV menjadi dataframe Pandas untuk pemrosesan lebih lanjut.

```python
f = pd.read_excel("Customer Call List.xlsx")
f
```

# Menghapus Kolom Tidak Berguna
- Menghapus kolom yang tidak diperlukan dalam analisis.

```python
f = f.drop_duplicates()
f
```

# Menghapus Kolom Tidak Berguna
- Menghapus kolom yang tidak diperlukan dalam analisis.

```python
f = f.drop(columns = "Not_Useful_Column")
f
```

# Menghapus Karakter Tertentu dari "Last_Name"
- Karakter `1`, `2`, `3`, `/`, `_` dihapus dari awal dan akhir string di kolom `Last_Name`.

```python
#f["Last_Name"] = f["Last_Name"].str.lstrip("...")
#f["Last_Name"] = f["Last_Name"].str.lstrip("/")
#f["Last_Name"] = f["Last_Name"].str.rstrip("_")
f["Last_Name"] = f["Last_Name"].str.strip("123._/")
f
```

# Mengubah Format "Phone_Number"
- Menghapus karakter non-alfanumerik dari kolom `Phone_Number`.
- Memformat nomor telepon menjadi format standar `xxx-xxx-xxxx`.
- Menghapus nilai `nan--` dan `Na--` yang mungkin muncul akibat data yang tidak valid.

```python
#f["Phone_Number"] = f["Phone_Number"].str.replace('[^a-zA-Z0-9]','')

#f["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#f["Phone_Number"] = f["Phone_Number"].apply(lambda x: str(x))

#f["Phone_Number"] = f["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

f["Phone_Number"] = f["Phone_Number"].str.replace('nan--','')

f["Phone_Number"] = f["Phone_Number"].str.replace('Na--','')
f
```

# Membagi Kolom "Address"
- Kolom `Address` dibagi menjadi tiga kolom baru: `Street_Address`, `State`, dan `Zip_Code` berdasarkan pemisah koma.

```python
f[["Street_Address", "State", "Zip_Code"]] = f["Address"].str.split(',',n=2, expand=True)
f
```

# Mengganti Nilai pada Kolom "Do_Not_Contact"
- Mengganti nilai `Yes` dengan `Y` dan `No` dengan `N` di kolom `Do_Not_Contact`.

```python
f["Do_Not_Contact"] = f["Do_Not_Contact"].str.replace('Yes','Y')

f["Do_Not_Contact"] = f["Do_Not_Contact"].str.replace('No','N')
f
```

# Mengganti Nilai yang Hilang
- Fungsi `fillna('')` digunakan untuk mengganti nilai yang hilang (NaN) dengan string kosong (`''`).

```python
#f = f.replace('N/a','')
#f = f.replace('NaN','')


f=f.fillna('')
f
```

# Menghapus Baris dengan Nilai 'Y' di "Do_Not_Contact"
- Baris yang memiliki nilai `Y` di kolom `Do_Not_Contact` akan dihapus.

```python
for x in f.index:
    if f.loc[x, "Do_Not_Contact"] == 'Y':
        f.drop(x, inplace=True)

f
```

# Menghapus Baris dengan Nilai Kosong di "Phone_Number"
- Looping digunakan untuk memeriksa setiap baris. Jika kolom `Phone_Number` kosong (`''`), maka baris tersebut dihapus.

```python
for x in f.index:
    if f.loc[x, "Phone_Number"] == '':
        f.drop(x, inplace=True)

f
```

# Mereset Indeks
- `reset_index(drop=True)` digunakan untuk mereset indeks pada dataframe dan menghilangkan kolom indeks lama.

```python
f = f.reset_index(drop=True)
f
```
